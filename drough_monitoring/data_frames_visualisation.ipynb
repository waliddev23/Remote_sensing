```python
# elaboration des grapphs pour vci tci pci spi 
# graph d'evolution de different indice suivant le temp , heatmaps pour les valeurs suivant le temp calcul de correlation 
#visualisation pour le vci 
import matplotlib.pyplot as pyplt
pyplt.rcParams["figure.figsize"] = (18, 10)
plt.style.use('fivethirtyeight')
# Assuming 'd' is your DataFrame
plt.figure(figsize=(70, 70))  # Increase the figure size
ax = vci.plot()
ax.set_xlabel('Mois', fontsize=18)
ax.set_ylabel('Vci_valeurs', fontsize=18)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title='Annee', fontsize ='xx-large' )  # Position the legend outside
ax.tick_params(axis='both', which='major', labelsize=16)
drought_classes = [
    (0, 10, 'extreme'),
    (10, 20, 'severe'),
    (20, 30, 'moderee'),
    (30, 40, 'moyenne'),
    (40, 100, 'aucune')
]
colors = ["purple" , "r" ,"orange", "y" , "g" ]
# Plot horizontal lines and text annotations for each drought class
index = 0 
for (ymin, ymax, label) in drought_classes:
    plt.axhspan(ymin, ymax, facecolor=colors[index], alpha=0.3)  # Adding shaded regions for clarity
    plt.text(0, (ymin + ymax) / 2, label, va='bottom', ha='center', fontsize=13, bbox=dict(facecolor='white', alpha=0.3))
    index = index +1 

# Adjust layout to make room for the legend
plt.tight_layout()

# Show plot
 # Adjust layout to make room for the legend
plt.show()
#fontsizeint or {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
```
